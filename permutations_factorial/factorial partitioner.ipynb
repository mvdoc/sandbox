```python
from itertools import *
```

```python
def get_unique_perms(perms):
    seen = set()
    for p in perms:
        # split it up into two groups
        plen = len(p)
        p_ = (p[:plen//2], p[plen//2:])
        p__ = tuple(sorted(tuple(map(lambda x: tuple(sorted(x)), p_))))
        if p__ not in seen:
            seen.add(p__)
    return list(seen)
```

```python
label2x2 = permutations(range(4), 4)
unique_perms_2x2 = get_unique_perms(label2x2)
print 'Got {0} unique perms'.format(len(unique_perms_2x2))
unique_perms_2x2
```
